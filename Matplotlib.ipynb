# ***Matplotlib***

```python
from numpy.lib.function_base import average
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


df = pd.read_csv('/content/Iris.csv')
df.head()

sl = df['SepalLengthCm']
sw = df['SepalWidthCm']
pl = df['PetalLengthCm']
pw = df['PetalWidthCm']
sps = df['Species']

plt.bar(1,sl, color = 'g')
plt.bar(2,sw, color = 'r')
plt.bar(3,pl, color = 'b')
plt.bar(4,pw, color = 'y')
plt.bar(5,sps, color = 'r')

plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```
